````python
# код з чату GPT

from typing import List

def calculate_variance(data: List[float]) -> float:
    if len(data) < 2:
        raise ValueError("Variance calculation requires at least two data points.")
    
    # Calculate the mean and mean of squares using built-in functions
    mean = sum(data) / len(data)
    mean_squares = sum(x ** 2 for x in data) / len(data)
    
    # Calculate variance using the formula: variance = mean(squares) - mean^2
    variance = mean_squares - mean ** 2
    
    return variance
````
