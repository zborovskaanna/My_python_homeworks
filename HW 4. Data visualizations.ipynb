__15.2__ Вивести графік динаміки поїздок по годинам

```python
df_taxi_clean['pickup_hour'] = df_taxi_clean['pickup'].dt.hour
```

```python
pivot_hour_pickup = df_taxi_clean[df_taxi_clean['pickup_borough'] != 'NA'].pivot_table(
    index='pickup_hour', columns='pickup_borough', aggfunc='size')
```

```python
import numpy as np
```

```python
pivot_hour_pickup.fillna(0).plot(figsize=(12,6))

plt.xticks(np.arange(24)) 
plt.grid(alpha=.3)

plt.title('Hourly dynamics of the number of rides')
plt.xlabel('Hours (pickup)')
plt.ylabel('Count of rides (total)');
```

Цей графік допоможе розуміти динаміку здійснення поїздок по годинам. Від цього регулювати кількість автівок, готових
приймати замовлення.
В районі Manhattan дуже активно користувались послугами таксі з 8-ї до 23-ї, пікові години - 17, 18, 19. 
В Bronх з 23-ї по 5-ту ранку або зовсім не було замовлень, або були одиничні. Загалом тут кількість поїздок

__15.3__ В якій кількості клієнти користуються платними дорогами в порівнянні з безкоштовними по кожному району?

```python
def tolls_to_category(float_toll):

    if float_toll == 0:
        return 'free'
    else:
        return 'paid'
```

```python
df_taxi_clean['tolls_catigorical'] = df_taxi_clean['tolls'].apply(tolls_to_category)
```

```python
sns.catplot(
    data=df_taxi_clean[(df_taxi_clean['pickup_borough'] != 'NA')], 
    x='pickup_borough', kind='count', hue='tolls_catigorical');
```

Серед усіх районів, клієнти з  Queens найбільше користуються платними дорогами. Такі поїздки становлять суттєву частину від усіх поїздок в цьому районі. Це варто врахувати при розрахунку тарифів, формуванні аційних пропозицій тощо.
